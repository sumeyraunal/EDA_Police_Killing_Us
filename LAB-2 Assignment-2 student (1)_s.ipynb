## 2: Plot 12 most dangerous cities

```python
df.head(1)
```

```python
df['city'].unique()
```

```python
df['city'].value_counts()[:12]
```

```python
plt.figure(figsize=(8,5))
sns.barplot(df['city'].value_counts()[:12].keys(),df['city'].value_counts()[:12])
plt.xticks(rotation=60)
plt.title('Most dangeours cities', color='blue', size=15);
```

```python
# with seaborn barplot

```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x = 'city',
              data = df,
              order = df['city'].value_counts().head(12).index)
plt.xticks(rotation=60)
plt.title("Most dangerous cities", color= 'b', size= 15 );
```

```python
# with seaborn countplot
```

## 3: Plot 20 most dangerous states
